```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn import svm
from sklearn.model_selection import train_test_split
from IPython.display import display, HTML, Image
import matplotlib.pyplot as plt
%matplotlib inline
```

Importing files containing data from: 
* Site data from Coollegrean 
* MERRA-2 data near the site (reference data from climate models)
* Valentia observatory data (reference data)

```python
#Site data is in two files hence reading and combining them into one
site_data_1 = pd.read_csv(r'/home/inder/Desktop/Thesis/Project_Inder/Coollegrean_file1.csv', usecols=['TMSTAMP','WS5_ms_Avg','WindDir2_AVG'],\
                          parse_dates=['TMSTAMP'], dayfirst=True)
site_data_2 = pd.read_csv(r'/home/inder/Desktop/Thesis/Project_Inder/Coollegrean_file2.csv', usecols=['TMSTAMP','WS5_ms_Avg','WindDir2_AVG'],\
                          parse_dates=['TMSTAMP'], dayfirst=True)
site_data = pd.concat([site_data_1, site_data_2])

site_data.to_csv(r'/home/inder/Desktop/Thesis/Project_Inder/site_data.csv')
#Dropping rows with 0 wind-speed
site_data = site_data.drop(site_data[site_data['WS5_ms_Avg']==0].index)
```

```python
#MERRA-2 data
merra2 = pd.read_csv(r'/home/inder/Desktop/Thesis/Project_Inder/merra2_coollegrean.csv', usecols=['DateTime','WS50m_m/s','WD50m_deg'],\
                     parse_dates=['DateTime'])
merra2 = merra2.set_index('DateTime')
```

```python

#Valentia Observatory data
valentia = pd.read_csv(r'/home/inder/Desktop/Thesis/Project_Inder/valentia_observatory.csv', skiprows=23, usecols=['date','wdsp','wddir'],\
                       parse_dates=['date'])
#Data is stored as string in columns we need to convert it into numbers
valentia['wdsp'] = pd.to_numeric(valentia['wdsp'],errors='coerce')
valentia['wddir'] = pd.to_numeric(valentia['wddir'],errors='coerce')
valentia = valentia.dropna()
valentia = valentia.set_index('date')
```

### The site_data is 10 minute but both the reference sets have hourly data. Hence we need to convert site_data to hourly_site_data by taking averages over each hour.

```python
hourly_site_data = site_data.groupby(pd.Grouper(freq='1H',key='TMSTAMP')).mean()
hourly_site_data = hourly_site_data.dropna()
```

```python
def overlap(df1, df2):
    """Finds overlapping indexes from two dataframes and also the percentage of indexes in df2.
    Coverage is 1 if whole site dat is covered.
    """
    ovlp = df1.index.intersection(df2.index)
    coverage = float(len(ovlp))/float(len(df2.index)) 
    return ovlp, coverage
```

```python
def timeseries_plot(overlap_idx,d1,d2, d1_label, d2_label):
    """For plotting time series for windspeed from two sites"""
    fig1 = plt.figure()
    ax = fig1.add_subplot(111)
    ax.grid()
    ax.set_ylabel('Wind Speed')
    ax.set_xlabel('Time')
    ax.set_title('Correlation')
    ax.plot(overlap_idx[-300:], d1[overlap_idx[-300:]], alpha=0.8,linewidth=2, label=d1_label)
    ax.plot(overlap_idx[-300:], d2[overlap_idx[-300:]],alpha=0.5, linewidth=2, label =d2_label)
    ax.legend(loc='upper left')
    fig1.set_figwidth(26)
    fig1.set_figheight(10)
 
```

```python
def scatter_plot(overlap_idx, x, y, x_label, y_label):
    fig2 = plt.figure(111)
    scat = fig2.add_subplot(111)
    scat.set_xlabel(x_label)
    scat.set_ylabel(y_label)
    scat.scatter(x[overlap_idx], y[overlap_idx])
    fig2.set_figwidth(10)
    fig2.set_figheight(10)
    
```

## Correlation MERRA-2 vs Coollegrean (hourly)

```python
#Finding overlapping indexes from both the datasets
overlap_idx, coverage = overlap(merra2, hourly_site_data)
print("Coverage:", coverage);

#Linear Regression
m_vs_c = LinearRegression()
m_vs_c.fit(merra2['WS50m_m/s'][overlap_idx].values.reshape(-1,1),hourly_site_data['WS5_ms_Avg'][overlap_idx].values.reshape(-1,1))
r2 = m_vs_c.score(merra2['WS50m_m/s'][overlap_idx].values.reshape(-1,1),hourly_site_data['WS5_ms_Avg'][overlap_idx].values.reshape(-1,1)) 
print("R2 coefficient:",r2)

#SVM
merra2_train, merra2_test, site_train, site_test = train_test_split(merra2['WS50m_m/s'][overlap_idx].values.reshape(-1,1),\
                                                    hourly_site_data['WS5_ms_Avg'][overlap_idx].values.reshape(-1,1), test_size=0.3, random_state=100)
svm_obj = svm.SVR()
svm_obj.fit(merra2_train, site_train)
prediction = svm_obj.predict(merra2_test)
plt.plot(prediction-site_test, 'ro')
plt.xlabel('Prediction')
plt.ylabel('Actual')
                                                                    
timeseries_plot(overlap_idx, merra2['WS50m_m/s'],hourly_site_data['WS5_ms_Avg'],'MERRA2', 'Site data')
scatter_plot(overlap_idx, merra2['WS50m_m/s'], hourly_site_data['WS5_ms_Avg'], 'MERRA2', 'Site data')
```

```python
overlap_idx = merra2.index.intersection(hourly_site_data.index)
check_merra2 = pd.concat([merra2['WS50m_m/s'][overlap_idx],hourly_site_data['WS5_ms_Avg'][overlap_idx] ], axis=1)
check_merra2.to_csv(r'/home/inder/Desktop/Thesis/Project_Inder/check_merra2.csv')
```

## Correlation Valentia Observatory vs Coollegrean (hourly)

```python
overlap_idx, coverage = overlap(valentia, hourly_site_data)
print("Coverage:", coverage);

v_vs_c = LinearRegression()
v_vs_c.fit(valentia['wdsp'][overlap_idx].values.reshape(-1,1),hourly_site_data['WS5_ms_Avg'][overlap_idx].values.reshape(-1,1))
r2 = v_vs_c.score(valentia['wdsp'][overlap_idx].values.reshape(-1,1),hourly_site_data['WS5_ms_Avg'][overlap_idx].values.reshape(-1,1)) 
print("R2 coefficient:",r2)

timeseries_plot(overlap_idx, valentia['wdsp'],hourly_site_data['WS5_ms_Avg'],'MERRA2', 'Site data')
scatter_plot(overlap_idx, valentia['wdsp'], hourly_site_data['WS5_ms_Avg'], 'MERRA2', 'Site data')
```

```python
overlap_idx = valentia.index.intersection(hourly_site_data.index)
check_valentia = pd.concat([valentia['wdsp'][overlap_idx],hourly_site_data['WS5_ms_Avg'][overlap_idx] ], axis=1)
check_valentia.to_csv(r'/home/inder/Desktop/Thesis/Project_Inder/check_valentia.csv')
```

## Correlation combined (Valentia Observatory - MERRA-2)  vs Coollegrean (hourly)


```python
overlap_idx = valentia.index.intersection(hourly_site_data.index)
overlap_idx = merra2.index.intersection(overlap_idx)

#To check if all the timestamps at site are in reference. Coverage is 1 if whole site dat is covered
print("Coverage:", float(len(overlap_idx))/float(len(hourly_site_data.index)))

combined = LinearRegression()
combined_data = pd.concat([valentia['wdsp'][overlap_idx],merra2['WS50m_m/s'][overlap_idx]], axis=1)
combined.fit(combined_data, hourly_site_data['WS5_ms_Avg'][overlap_idx].values.reshape(-1,1))
r2 = combined.score(combined_data, hourly_site_data['WS5_ms_Avg'][overlap_idx].values.reshape(-1,1))
print("R2 coefficient:",r2)
```

```python
overlap_idx = valentia.index.intersection(hourly_site_data.index)
overlap_idx = merra2.index.intersection(overlap_idx)
check_both = pd.concat([merra2['WS50m_m/s'][overlap_idx], valentia['wdsp'][overlap_idx],hourly_site_data['WS5_ms_Avg'][overlap_idx] ], axis=1)
check_both.to_csv(r'/home/inder/Desktop/Thesis/Project_Inder/check_both.csv')
```
